```python
# Importar librerias
import matplotlib.pyplot as plt
import pandas as pd
```

Leer csv

```python
df = pd.read_csv('/content/drive/MyDrive/BOOTCAMP UTB/Agricultura_y_Desarrollo_Rural.csv')
df
```

```python
# Tabla Cruzada
tabla_cruzada = pd.crosstab(df['Municipio'], df['Cultivo'])
tabla_cruzada
```

```python
#Grafico de barras apiladas
fig,ax =plt.subplots(figsize=(16,10))
tabla_cruzada.plot(kind='bar', stacked=True, ax=ax)
ax.legend(title='Cultivos', bbox_to_anchor=(1.05, 1), loc='upper left', ncol=3)
plt.show()
```

Limpiar datos (Colocar en minusculas)

```python
df['Municipio'] = df['Municipio'].str.lower().str.strip()
df['Cultivo'] = df['Cultivo'].str.lower().str.strip()
```
